# Adversarial Test Demo

Detect drift between two datasets using adversarial validation.

```python
import numpy as np
import pandas as pd
from caketool.monitor import AdversarialModel
```

## Create Sample Data

```python
np.random.seed(42)
n = 1000

# Reference dataset (training data)
df_train = pd.DataFrame({
    "age": np.random.normal(35, 10, n),
    "income": np.random.normal(50000, 15000, n),
    "category": np.random.choice(["A", "B", "C"], n),
    "region": np.random.choice(["North", "South", "East", "West"], n),
})

df_train.head()
```

## Scenario 1: No Drift

New data follows same distribution as training data.

```python
np.random.seed(123)

# New dataset with SAME distribution
df_no_drift = pd.DataFrame({
    "age": np.random.normal(35, 10, n),
    "income": np.random.normal(50000, 15000, n),
    "category": np.random.choice(["A", "B", "C"], n),
    "region": np.random.choice(["North", "South", "East", "West"], n),
})

model = AdversarialModel()
model.fit(df_train, df_no_drift)
model.show()
```

**AUC ~0.5**: Model cannot distinguish between datasets → No drift detected

## Scenario 2: With Drift

New data has different distribution (population shift).

```python
np.random.seed(456)

# New dataset with DRIFT
df_drift = pd.DataFrame({
    "age": np.random.normal(45, 10, n),      # older population
    "income": np.random.normal(70000, 15000, n),  # higher income
    "category": np.random.choice(["A", "B", "D"], n),  # new category D
    "region": np.random.choice(["North", "South"], n),  # only 2 regions
})

model = AdversarialModel()
model.fit(df_train, df_drift)
model.show()
```

**AUC > 0.7**: Model can distinguish datasets → Drift detected!

Top features show which variables contributed most to the drift.

## Interpretation

| AUC Score | Interpretation |
|-----------|----------------|
| ~0.5 | No drift (random guessing) |
| 0.5 - 0.7 | Mild drift |
| > 0.7 | Significant drift |
| > 0.9 | Severe drift |
